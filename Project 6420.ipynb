```python
from flask import Flask, render_template_string, request, redirect, url_for
from pymongo import MongoClient
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)  # Enables ngrok for public access

# MongoDB configuration
MONGO_URI = "mongodb://localhost:27017/"
client = MongoClient(MONGO_URI)
db = client["data_collection_db"]
collection = db["user_data"]

# HTML form as a template string
form_html = """
<!DOCTYPE html>
<html>
<head>
    <title>Data Collection Form</title>
</head>
<body>
    <h1>Data Collection Form</h1>
    <form action="/submit" method="post">
    <label for="name">Name:</label>
    <input type="text" id="name" name="name" required><br><br>
    
    <label for="age">Age:</label>
    <input type="number" id="age" name="age" required><br><br>
    
    <label for="gender">Gender:</label>
    <select id="gender" name="gender" required>
        <option value="Male">Male</option>
        <option value="Female">Female</option>
        <option value="Other">Other</option>
    </select><br><br>
    
    <label for="expenses">Expenses:</label>
    <input type="number" id="expenses" name="expenses" required><br><br>
    
    <label for="total_income">Total Income:</label>
    <input type="number" id="total_income" name="total_income" required><br><br>
    
    <button type="submit">Submit</button>
</form>
</body>
</html>
"""

@app.route('/')
def index():
    return render_template_string(form_html)

@app.route('/submit', methods=['POST'])
def submit():
    if request.method == 'POST':
        # Collect data from the form
        name = request.form.get('name')
        age = int(request.form.get('age'))
        gender = request.form.get('gender')
        expenses = float(request.form.get('expenses'))
        total_income = float(request.form.get('total_income'))
        
        

        # Create a document to insert into MongoDB
        data = {
            "name": name,
            "age": age,
            "gender": gender,
            "expenses": expenses,
            "total_income": total_income
            }

        try:
            # Insert data into the MongoDB collection
            collection.insert_one(data)
        except Exception as e:
            return f"An error occurred while submitting the data: {e}"

        return redirect(url_for('success'))

@app.route('/success')
def success():
    return "Data submitted successfully!"

if __name__ == '__main__':
    app.run()
```

```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Load the dataset
file_path = r"C:\Users\25677\Project Data.csv"
data = pd.read_csv(file_path)

# Sort the data by "Total Income" in descending order
sorted_data = data.sort_values(by="Total Income", ascending=False)

# Aggregate incomes for duplicate ages and get the top 10
aggregated_data = sorted_data.groupby("Age", as_index=False).sum()
top_ages = aggregated_data.sort_values(by="Total Income", ascending=False).head(10)

# Create a data source for Bokeh
source = ColumnDataSource(top_ages)

# Create a bar chart
p = figure(
    x_range=source.data["Age"].astype(str),
    title="Top Ages Based on Total Income",
    x_axis_label="Age",
    y_axis_label="Total Income",
    width=800,
    height=400
)
p.vbar(
    x="Age",
    top="Total Income",
    width=0.9,
    source=source,
    color="dodgerblue"
)

# Customize the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
output_notebook()
show(p)
```
